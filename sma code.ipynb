```python
import pandas as pd 
import matplotlib.pyplot as plt 
import requests
import math
from termcolor import colored as cl 
import numpy as np
```

```python
msft = pd.read_csv('RELIANCE.csv').set_index('Date')
msft.index = pd.to_datetime(msft.index)
msft.tail()
```

```python
def sma(data, n):
    sma = data.rolling(window = n).mean()
    return pd.DataFrame(sma)

n = [20, 50]
for i in n:
   msft[f'sma_{i}'] = sma(msft['Close'], i)
msft.tail()
```

```python
from matplotlib.figure import Figure

plt.plot(msft['Close'], label = 'MSFT' , color='skyblue')
plt.plot(msft['sma_20'], label = 'SMA 20', color='red')
plt.plot(msft['sma_50'], label = 'SMA 50',color='yellow')
plt.title('MSFT Simple Moving Averages (20, 50)')
plt.legend(loc = 'upper left')
fig = plt.gcf()
fig.savefig('smaASIAN.png', dpi=150)
plt.show()
plt.draw()
```
